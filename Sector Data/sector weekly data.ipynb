```python
XLI = pd.read_csv('weekly_XLI.csv')
XLE = pd.read_csv('weekly_XLE.csv')
XLB = pd.read_csv('weekly_XLB.csv')
XLY = pd.read_csv('weekly_XLY.csv')
XLP = pd.read_csv('weekly_XLP.csv')
XLV = pd.read_csv('weekly_XLV.csv')
XLF = pd.read_csv('weekly_XLF.csv')
XLK = pd.read_csv('weekly_XLK.csv')
XLC = pd.read_csv('weekly_XLC.csv')
XLU = pd.read_csv('weekly_XLU.csv')
XLRE = pd.read_csv('weekly_XLRE.csv')
```

```python
sectors = [XLI, XLE, XLB, XLY, XLP, XLV, XLF, XLK, XLC, XLU, XLRE]
new_sectors = []
for i in sectors:
    i["timestamp"] = pd.to_datetime(i["timestamp"])
    i = i.loc[(i['timestamp'] > '2010-01-01')]
    new_sectors.append(i)
new_sectors
```

```python
import plotly.graph_objects as go
close_price = go.Figure()
name_sectors = ['XLI', 'XLE', 'XLB', 'XLY', 'XLP', 'XLV', 'XLF', 'XLK', 'XLC', 'XLU', 'XLRE']
for i in range(0, len(new_sectors)):
    close_price.add_trace(go.Scatter(x=new_sectors[i]['timestamp'], y=new_sectors[i]['close'], name=name_sectors[i]))

close_price
```

```python
low_price = go.Figure()
name_sectors = ['XLI', 'XLE', 'XLB', 'XLY', 'XLP', 'XLV', 'XLF', 'XLK', 'XLC', 'XLU', 'XLRE']
for i in range(0, len(new_sectors)):
    low_price.add_trace(go.Scatter(x=new_sectors[i]['timestamp'], y=new_sectors[i]['low'], name=name_sectors[i]))

low_price
```
