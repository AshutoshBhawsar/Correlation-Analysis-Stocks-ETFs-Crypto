```python
# Correlation Moving Average Function
# Confirm the input DF index is a Datetime index

def corrMA(df, window="30D", test="btc-usd", corrs=["eth-usd"]):
    temp = df.rolling(window).corr()
    
    # Insert the independent variable ticker to the corrs list
    if not test in corrs:
        corrs.insert(0, test)
        
    df_corr = temp[test].unstack()[corrs]
    # return temp[test].unstack()
    
    title = "Correlations of: " + str(corrs) + " relative to " + test
    df_corr.plot(title=title)
    
    df_corr.rename(columns={test: "corr-"+test}, inplace=True)
    for i in corrs:
        df_corr.rename(columns={i: "corr-"+i}, inplace=True)
        
    return df_corr

btc_eth_30 = corrMA(crypto_df_dropna)
btc_eth_30

# crypto_df_dropna.rolling("30D").corr()
```

# Function for corr & plots of situations
### Also, checking the pricing for stuff

- df = dataframe with prices
- window = INTEGER representing the time window in DAYS --> 30, 50, 200, etc. (or some other valid window string)
- independent = string of indp. variable ticker (dominant asset i.e. btc)
- dependent = 1 dep. variable ticker(s) (dependent asset i.e. eth)
- corr_ratio = float for correlation ratio (.3, .5, etc.)

```python
from datetime import timedelta
from IPython.display import display, HTML

def corrma_analysis(df, window, independent, dependent, corr_ratio):
    # Get the overall corrma for the window
    corr_df = corrMA(df, str(window) + "D", independent, [dependent])
    # print(corr_df.where(corr_df["corr-"+dependent] < corr_ratio).dropna())
    # print(corr_df)
    
    '''I am missing some data here. There is a part of the window at the start of this timeframe which is dropped for some reason.'''
    '''Added a thresh=1 parameter to .dropna() below. This maintains the top rows (which might be dropped for some other reason).'''
    
    # Compute the price change over the window (observation purposes)
    price_change = df[[independent, dependent]].rolling(window).apply(lambda x: x.iloc[-1] - x.iloc[0])
    price_change.rename(columns={independent: "window-change-"+independent, dependent: "window-change-"+dependent}, inplace=True)
    
    # Merge the information together(corrs, price change, asset actual price)
    main_df = corr_df.join(price_change)
    main_df = main_df.join(df[[independent, dependent]])
    
    
    # Get the "low_corrs" based on the corr_ratio
    low_corrs = main_df.where(main_df["corr-" + dependent] < corr_ratio).dropna(thresh=1)  # .dropna() required to remove the other NaN rows
    # print(low_corrs)
    
    # Merge in these rows... Add a new column with the corr + NaN vals...
    main_df["corr-below-ratio"] = low_corrs[["corr-" + dependent]]
    
    # Add % change columns (for use if we want)
    main_df["%-" + independent] = main_df[independent].pct_change(1)
    main_df["%-" + dependent] = main_df[dependent].pct_change(1)
    
    # Compute cumulative returns
    main_df["cumret-"+independent] = main_df[[independent]].apply(lambda x: x / x[0])
    main_df["cumret-"+dependent] = main_df[[dependent]].apply(lambda x: x / x[0])
    
    
    # Test plots
    main_df.plot(y=[dependent, independent], title=dependent + " price vs. " + independent)
    
    main_df.plot(y=["cumret-"+dependent, "cumret-"+independent], 
                 title=dependent + " returns vs. " + independent).axhline(y = 1, color = "black", lw=2)
    # main_df.plot(y=["corr-"+dependent, "corr-"+independent])
    # main_df.plot(y=["corr-below-ratio", "corr-"+independent])
    
    
    # Get the WINDOW of days before/after the ratio drops below the threshold
    dates = list(main_df.dropna().index)

    day = dates[0]  # first day where corrma triggered

    joiner = main_df.loc[day - timedelta(days=window): day + timedelta(days=window)]  # initial rows for the index. 

    # For all dates where corrma ratio < threshold, find WINDOW days before and after it.
    for i in dates:
        temp = main_df.loc[i - timedelta(days=window): i + timedelta(days=window)]
        joiner = pd.concat([joiner, temp]).drop_duplicates()
        
    
    # Add a column to signify the special dates in main_df
    joiner_new = pd.DataFrame(index = joiner.index)
    joiner_new["adjusted"] = True

    main_df["dep-window-dates"] = joiner_new[["adjusted"]]
    
    
    # Filter out and plot the ADJUSTED DATES cumulative returns ! To show the gaps!
    # This will grab the dates found in the WINDOW period before and after the corrma threshold and plot the cumret
    range_df = main_df.where(main_df["dep-window-dates"])
    
    ax = range_df.plot.line(y=["cumret-"+dependent, "cumret-"+independent], title="Returns w/ periods of low correlation highlighted")
    
    # Split dates into VALID < corrma AND leave the WINDOWS alone (highlighting)
    
    # print(main_df.loc[main_df["corr-below-ratio"].isnull()])
    
    null_index = main_df.loc[main_df["corr-below-ratio"].isnull()].index
    valid_index = main_df.loc[main_df["corr-below-ratio"].notnull()].index

    null_dict = [{"date": i, "valid": False} for i in null_index]
    valid_dict = [{"date": i, "valid": True} for i in valid_index]

    null_df = pd.DataFrame(null_dict).set_index("date")
    null_df.index = pd.to_datetime(null_df.index)
    valid_df = pd.DataFrame(valid_dict).set_index("date")
    valid_df.index = pd.to_datetime(valid_df.index)


    x = pd.concat([null_df, valid_df])
    x.sort_index(inplace=True)

    # Working
    s = x["valid"].rolling(2).apply(lambda y: y[0] != y[1])
    s = s.fillna(0.0)
    changes = s.loc[s != 0.0]
    dates = changes.index
    dates

    # Add in the highlights on the chart :D (need to account for if the trend of < .3 corr is STILL HAPPENEING -> odd # elems)
    for i in range(len(dates)):
        if i % 2 == 0:
            start = i 
            end = i+1

            # i.e., if end DNE in the dates, the range HAS NOT ENDED YET (select the LAST elem)
            if end >= len(dates):
                end = main_df.index[-1]
                ax.axvspan(dates[start], end, facecolor="red", alpha=.2)
                
            else:
                ax.axvspan(dates[start], dates[end], facecolor="red", alpha=.2)


    # Re-plot the correlations chart w/ highlights
    ax2 = main_df.plot.line(y=["corr-below-ratio", "corr-"+independent], title="Correlations w/ periods of low correlation shown")

    for i in range(len(dates)):
        if i % 2 == 0:
            start = i 
            end = i+1

            # i.e., if end DNE in the dates, the range HAS NOT ENDED YET (select the LAST elem)
            if end >= len(dates):
                end = main_df.index[-1]
                ax2.axvspan(dates[start], end, facecolor="red", alpha=.2)
                
            else:
                ax2.axvspan(dates[start], dates[end], facecolor="red", alpha=.2)
            
    
    # Compute/showcase price details for each instance of low correlation
    # 1) Price WINDOW days (or as far as possible) before the occurrence
    # 2) Price at the occurence
    # 3) Price at the end of the occurrence
    # 4) Price WINDOW days past the occurrence (or as far as possible)
    
    # Upon each occurrence, do some statistics! How often? Up or down? What's going on?
    stats_df = pd.DataFrame(columns=["start", "end"])
    
    for i in range(len(dates)):
        if i % 2 == 0:
            start = i 
            end = i+1

            # i.e., if end DNE in the dates, the range HAS NOT ENDED YET (select the LAST elem)
            if end >= len(dates):
                end = main_df.index[-1]
                end_date = end
            else:
                end_date = dates[end]
                
            start_date = dates[start]

            # Computations
            stats = {}
            stats["start"] = [start_date]  # The Value has to be inside a [] for Pandas...
            stats["end"] = [end_date]
            
            # Pre-date
            try:
                pre = main_df.loc[start_date - timedelta(days=window)]
            except:
                pre = main_df.iloc[0]
                
            # Post-date
            try:
                post = main_df.loc[end_date + timedelta(days=window)]
            except:
                post = main_df.iloc[main_df.shape[0] - 1]

#             print(start_date)
#             print(pre, "\n")
#             print(end_date)
#             print(post, "\n")

            # Start Day CORRMA threshold
            s = main_df.loc[start_date]
            
            # End Day CORRMA threshold
            e = main_df.loc[end_date]
            

            # Prices
            stats["pre-ind-price"] = pre.loc[independent]
            stats["start-ind-price"] = s.loc[independent]
            stats["end-ind-price"] = e.loc[independent]
            stats["post-ind-price"] = post.loc[independent]
            
            stats["pre-dep-price"] = pre.loc[dependent]
            stats["start-dep-price"] = s.loc[dependent]
            stats["end-dep-price"] = e.loc[dependent]
            stats["post-dep-price"] = post.loc[dependent]
            
            # Cumulative returns
            stats["pre-ind-cumret"] = pre.loc["cumret-" + independent]
            stats["start-ind-cumret"] = s.loc["cumret-" + independent]
            stats["end-ind-cumret"] = e.loc["cumret-" + independent]
            stats["post-ind-cumret"] = post.loc["cumret-" + independent]
            
            stats["pre-dep-cumret"] = pre.loc["cumret-" + dependent]
            stats["start-dep-cumret"] = s.loc["cumret-" + dependent]
            stats["end-dep-cumret"] = e.loc["cumret-" + dependent]
            stats["post-dep-cumret"] = post.loc["cumret-" + dependent]
            
            stats["window-ind-%-change"] = (stats["end-ind-price"] - stats["start-ind-price"]) / stats["start-ind-price"]
            stats["window-dep-%-change"] = (stats["end-dep-price"] - stats["start-dep-price"]) / stats["start-dep-price"]
            
            # todo - Right now, I'm not seeing much of a pattern for the CORRMA thresholds being reached... 
            # for instance, there's no clear pattern that when the two diverge, that the DEP always increases, etc.
            # in fact, this could not have been hypothesized or assumed, as a lack of correlation simply means the two
            # assets are DIFFERENT. We do NOT know HOW they are different. One could go up, and the other could go up less.
            # Or, one could go up, the other could go down. 
            
            # todo - Find the LOWEST points of correlation for each range. And compare the situations where 
            # (a) correlation simply drops below the threshold BUT DOES NOT GO NEGATIVE
            # (b) correlation goes negative
            
            # Maybe there is something here!
        
        
            stats_df = pd.concat([stats_df, pd.DataFrame.from_dict(stats)])

    
    # Reset the index and present the data
    stats_df = stats_df.reset_index(drop=True)
    
    display(stats_df) # Imported method 
    # display(HTML(stats_df.to_html()))

    
    
    return main_df
```

* `window-change-<DEPENDENT>` - ^^
* `<INDEPENDENT>` - price of asset
* `<DEPENDENT>` - ^^
* `corr-below-ratio` - **column representing the corrma WHEN it crosses below the corr_ratio threshold**
* `%-<INDEPENDENT>` - % change per 1 day of asset
* `%-<DEPENDENT>` - ^^
* `cumret-<INDEPENDENT>` - cumulative returns of asset
* `cumret-<DEPENDENT>` - ^^
* `dep-window-dates` - **valid dates where it is within +/- 30 days of a valid "corr-below-ratio" day**
